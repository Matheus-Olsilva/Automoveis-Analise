```python
%%sql
SELECT Rank, count(Location_Region) as Contagem, Location_Region, Name, Company_Name, Year
FROM Bilionarios
Group by Location_Region, rank
Having Rank <=3
Order by Rank
```

```python
%%sql

CREATE TABLE Ranking4 AS
SELECT Rank, count(Location_Region) as Contagem, Location_Region, Name, Company_Name, Year
FROM Ranking3
GROUP BY Location_Region, Rank
HAVING Rank <= 3
ORDER BY Rank;
```

```python
%%sql
Select * From Ranking4
```
